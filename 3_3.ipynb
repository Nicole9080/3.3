```python
# Задание task_03_03_04.
#
# Выполнил: Манидарова Н.Ч.
# Группа: ЦИБ-251


"""
Ошибки (номера строк через пробел, данная строка - №2): 12 14
"""

def min_pair(nums):
    if not isinstance(nums, list):
        print("Ошибка: Переданный объект не является списком!")
        return None

    if len(nums) < 2:
        print("Ошибка: В списке должно быть минимум 2 числа!")
        return None

    min_sum = nums[0] + nums[1]

    for i in range(1, len(nums) - 1):
        current_sum = nums[i] + nums[i + 1]
        if current_sum < min_sum:
            min_sum = current_sum

    return min_sum

if __name__ == "__main__":
    import random

    random.seed(50)
    nums = random.sample(range(1, 100), 10)

    print("Список:", nums)
    print("Минимальная сумма соседних пар:", min_pair(nums))
```
